```python
import pandas as pd
```

```python
puma = pd.read_csv("california_pumacodes.csv")
puma
```

```python
# Remove the first 3 letters from the column "puma_code"
puma["puma_code"] = puma["puma_code"].str[3:]
puma["puma_int"] = puma["puma_code"].astype(int)
```

```python
puma
```

```python
# Get San Francisco PUMA codes

# Note that we're using "San Francisco County" to avoid matching the city of South San Francisco
sf_puma = puma[puma["puma_area_name"].str.contains("San Francisco County")]
sf_puma
```

```python
# Get LA PUMA codes

# Also using "County" suffix for consistency
la_puma = puma[puma["puma_area_name"].str.contains("Los Angeles County")]
la_puma
```

```python
# Write shorthands for the two regions
assert pd.Series.all(sf_puma["puma_int"] == list(range(7501, 7507+1)))
assert pd.Series.all(la_puma["puma_int"] == list(range(3701, 3769+1)))
```
